```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
# from datasets import load_dataset, load_metric
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Load dataset (your chatbot dataset)
dataset = load_dataset('csv', data_files='/content/healthcare_chatbot_dataset.csv')

# Extract unique intent labels from the "Intent" column
intent_labels = df['Intent'].unique().tolist()

# Initialize BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=len(intent_labels))  # num_labels depends on your intents

text_columns = ['Diet_Type', 'Health_Condition', 'Fitness_Goal', 'Food_Allergies', 'Meal_Preference', 'Cuisine_Type', 'Activity_Level', 'Snack_Type', 'Current_Habit', 'New_Habit', 'Preferred_Activity', 'Motivational_Trigger', 'Current_Routine', 'Optimized_Routine', 'Condition', 'Symptoms', 'Specialist_Type', 'Topic', 'User_Feedback']

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples['text'], padding="max_length", truncation=True)

# Apply tokenization only to the relevant columns
def tokenize_columns(examples):
    # Tokenize each text column and store the results in a dictionary
    tokenized_examples = {}
    for col in text_columns:
        tokenized_examples[col] = tokenizer(examples[col], padding="max_length", truncation=True)
    return tokenized_examples

tokenized_datasets = dataset.map(tokenize_columns, batched=True)

# Split dataset into train and validation
train_dataset, val_dataset = train_test_split(tokenized_datasets['train'], test_size=0.1)

# Define your custom metric function
def compute_metrics(p):
    preds, labels = p
    preds = preds.argmax(axis=-1)
    return {"accuracy": accuracy_score(labels, preds)}

# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
)

# Define Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)

# Train the model
trainer.train()
trainer.evaluate()


# Save the model
model.save_pretrained('./intent_recognition_model')
tokenizer.save_pretrained('./intent_recognition_model')
```
